```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Método de RKF e construa os gráficos das quantidades x1(t) e x2(t) 
# dados: x1(0) = 100, x2(0) = 10
# k12 = 0.03 , k21 = 0.02 e k20 = 0.01 
# x3(0) = 20 e k31 = 0.08
# gráficos x1(t), x2(t) e x3(t)
```

```python
def fx1(x1,x2):
    return 0.02*x2-0.03*x1
```

```python
def fx2(x1,x2):
    return 0.03*(x1 - x2)
```

```python
def RK4(h,f,x1,t):
    k1 = h*f(x1, t)
    k2 = h*f(x1 + 0.5*h, t + 0.5*k1)
    k3 = h*f(x1 + 0.5*h, t + 0.5*k2)
    k4 = h*f(x1 + h, t + k3)
    return x1 + (k1 /6 + k2/3 + k3/3 + k4 / 6)
```

```python
def RKF(x1,x2,v):
    h = 0.001
    t = 0.1
    while x2[len(x2)-1]>0:
        x1.append(RK4(h,fx1,x1[len(x1)-1],t))
        x2.append(RK4(h,fx2,x2[len(x2)-1],t))
        t += h
        v.append(t)
```

```python
x1, x2, v = [100], [10], [0]
RKF(x1,x2,v)
print("x1\tx2")

for i in range(len(x2)):
    print(round(x1[i],2),"\t",round(x2[i],2),"\t")
```

```python
plt.plot(v,x1)
plt.plot(v,x2)
plt.show()
```
